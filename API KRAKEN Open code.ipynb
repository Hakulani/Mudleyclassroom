```python
#Import library
import ccxt
import json
import pandas as pd
```

```python
# User Api Key and secret Key
import ccxt
exchange = ccxt.kraken ({
    'apiKey': 'YOURAPIKEY',
    'secret': 'YOURsecret',
    'enableRateLimit': True,
})
#Check  PORTFOLIO BALANCE 
print("PORTFOLIO BALANCE  =",exchange.fetch_balance())
```

```python
#The getattr() method returns the value of the named attribute of an object. If not found, it returns the default value provided to the function
#Check API function
id = 'kraken'
exchange = getattr(ccxt, id) ()
print(exchange.has)
```

```python
# Price product  By json
r1 = json.dumps(exchange.fetch_ticker('BTC/USD'))
r2 = json.dumps(exchange.fetch_ticker('XRP/USD'))
dataPriceBTC = json.loads(r1)
dataPriceXRP = json.loads(r2)

# Print Price
print(exchange)
print('BTC/USD =',dataPriceBTC['last'])
print('XRP/USD =',dataPriceXRP['last'])
```

```python
#  Check PRODUCT PAIR TRADE
print("PRODUCT =",exchange.symbols)
```

```python
# Place OPEN/CLOSE ORDER
#ordertype = order type:
#lmt for a limit order
#post for a post-only limit order
#stp for a stop order
#take_profit for a take profit order
#ioc for an immediate-or-cancel order

pair = "BTC/USD"      # PRODUCT PAIR
types = 'limit'       # type 
side = 'buy'          # buy/sell
size_order =  50      # @param {type:"number"}
price = 0.1770        # Order Price 
reduceOnly = False    # reduceOnly 	Set as true if you wish the order to only reduce an existing position. Any order which increases an existing position will be rejected. Default false.
postOnly =  False    
ioc = False           

exchange.create_order(pair, types , side, size_order, price) #Place order to exchange
```

```python
# Cancel PENDING ORDER
exchange.cancel_order('id เอามาจากตารางด้านล่าง ที่เราใช้ตรวจเช็คออเดอร์')
```

```python
#Check pending order
df1 = pd.DataFrame(exchange.fetch_open_orders(pair))
display(df1)
```

```python
# Check CLOSED or CANCELED
df3 = pd.DataFrame(exchange.fetch_closed_orders(pair))
display(df3)
```

```python
# FILTER CLOSED Only
pd5 = pd.DataFrame(filter(lambda x:x['status'] == "closed",exchange.fetch_closed_orders(pair)),
                   columns=['id','datetime','status','symbol','type','side','price','amount','filled','average','remaining'])
display(pd5)
```

```python
 #FILTER CANCELED Only
pd5 = pd.DataFrame(filter(lambda x:x['status'] == "canceled",exchange.fetch_closed_orders(pair)),
                   columns=['id','datetime','status','symbol','type','side','price','amount','filled','average','remaining'])

display(pd5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
sns.set_style("whitegrid")


timeframe = "5m" #@param ["1m", "5m", "15m", "1h", "4h", "1d"]
limit =  5000        #@param ["5000"] {type:"raw", allow-input: true}
ohlcv  = exchange.fetch_ohlcv(pair,timeframe , limit=limit )
ohlcv = exchange.convert_ohlcv_to_trading_view(ohlcv)
df =  pd.DataFrame(ohlcv)
```
